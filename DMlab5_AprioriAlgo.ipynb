```python
# 9. Create a transaction table in the format reqiured for Apriori Alogorithm calculate support,
# confidence and lift for single item dataset count no of k-itemset consider all the frequent itemsets for rule
# generation  where "lift">=1 which is the most frequently purchased product(x) and what are the other product
# purchased with X where "lift">=1
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

transactions = df.groupby("Transaction")["Item"].apply(list).tolist()

te = TransactionEncoder()
te_ary = te.fit_transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns = te.columns_)

frequent_itemsets = apriori(df_encoded, min_support=0.01, use_colnames=True)

single_itemset = frequent_itemsets[frequent_itemsets["itemsets"].apply(lambda x:len(x)==1)]
single_itemset["support"] = single_itemset["support"] * len(transactions)

k_itemsets_counts = frequent_itemsets["itemsets"].apply(lambda x:len(x))

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

most_frequent_product = frequent_itemsets.loc[frequent_itemsets["support"].idxmax(), "itemsets"]
related_products = rules.loc[(rules["antecedents"] == most_frequent_product) & (rules["lift"] >= 1), "consequents"].copy()
# related_products = rules[(rules["antecedents"] == most_frequent_product) & (rules["lift"] >=1)]["consequents"]

print("Transaction Table: ")
print(df_encoded.head())
print()
print("Support, Confidence and Lift for Single Item Dataset: ")
print(single_itemset)
print()
print("Number of K-itemsets: ")
print(k_itemsets_counts.value_counts())
print()
print("Association Rules: ")
print(rules)
print()
print("Most Frequently Purchased Product:")
print(most_frequent_product)
print()
print("Other Products Purchased with the most frequent Product(Lift>=1): ")
print(related_products)
```
